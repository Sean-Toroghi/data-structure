## <a name = 'array'>Array</a>

An Array is a collection of similar data types stored in contiguous memory locations. Array is a static data structure (it has a fixed size), and stores two additional information: _size_ and _capacity_.

__Array is a good data structure when we have a static data.__

#### B-trees

Summary
- a type of self-balancing tree
- also known as “large key” trees b/c of the large number of keys that they can store in a single node
- Each node in a B-Tree can contain multiple keys, which allows the tree to have a larger branching factor and thus a shallower height. -> less disk I/O, which results in faster search and insertion operations
- B-Trees maintains balance by ensuring that each node has a minimum number of keys, so the tree is always balanced.
- Each node can have at most 2t-1 keys and 2t children. The root can have at least one key and at most 2t-1 keys. All non-root nodes have at least t-1 keys and at most 2t-1 keys.
- B-Trees are optimized for the storing and retrieval of massive volumes of data from relatively slow storage media such as discs
- nodes are sorted in the inorder traversal
- each node could have more than two children

Operations:

| Algorithm | Time Complexity |
|:---------:|:---------------:|
|   Search  |     O(log n)    |
|   Insert  |     O(log n)    |
|   Delete  |     O(log n)    |

####  Red-black trees

Summary

- a type of balanced binary search tree that use a specific set of rules to ensure that the tree is always balanced.
- in addition to BST properties, it satisfies the following properties:
  1. Root property: The root is black.
  2. External property: Every leaf (Leaf is a NULL child of a node) is black
  3. Internal property: The children of a red node are black. Hence possible parent of red node is a black node.
  4. Depth property: All the leaves have the same black depth.
  5. Path property: Every simple path from root to descendant leaf node contains same number of black nodes.
- rules:
  - Every node has a color either red or black.
  - The root of the tree is always black.
  - There are no two adjacent red nodes (A red node cannot have a red parent or red - child).
  - Every path from a node (including root) to any of its descendants NULL nodes has the same number of black nodes.
  - Every leaf (e.i. NULL node) must be colored BLACK.

#### AVL tree

Summary
- AVL tree is a self-balancing Binary Search Tree (BST) where the difference between heights of left and right subtrees cannot be more than one for all nodes
-  AVL trees are more balanced compared to Red-Black Trees, but they may cause more rotations during insertion and deletion.
- Advantages Of AVL Trees:
  - It is always height balanced
  - Height Never Goes Beyond LogN, where N is the number of nodes
  - It give better search than compared to binary search tree
  - It has self balancing capabilitie
- Insert, delete, and seach in $O(\log n)$ time
- AVL tree are mostly used where search is more frequent compared to insert and delete operation.


__Implementation__
- Insert: insert using BST steps. Then apply right or left rotation until reach AVL property.
- Delete: first perform standart BST deletion steps. To balance the tree follow the following steps:
  - Starting from w, travel up and find the first unbalanced node. Let z be the first unbalanced node, y be the larger height child of z, and x be the larger height child of y. Note that the definitions of x and y are different from insertion here.
  -  Re-balance the tree by performing appropriate rotations on the subtree rooted with z. There can be 4 possible cases that needs to be handled as x, y and z can be arranged in 4 ways. Following are the possible 4 arrangements:
    - y is left child of z and x is left child of y (Left Left Case)
    - y is left child of z and x is right child of y (Left Right Case)
    - y is right child of z and x is right child of y (Right Right Case)
    - y is right child of z and x is left child of y (Right Left Case)

####  Splay trees

Summary
- a self-adjusting binary search tree data structure
- The basic idea behind splay trees is to bring the most recently accessed or inserted element to the root of the tree by performing a sequence of tree rotations, called splaying. Splaying is a process of restructuring the tree by making the most recently accessed or inserted element the new root and gradually moving the remaining nodes closer to the root.
- :(  main disadvantage of splay trees is that they do not guarantee a balanced tree structure
- each time an element is accessed, it is moved to the root of the tree, creating a more balanced structure for subsequent accesses.

- Applications of the splay tree:
  - Caching: Splay trees can be used to implement cache memory management, where the    most frequently accessed items are moved to the top of the tree for quicker access.
  - Database Indexing: Splay trees can be used to index databases for faster searching  and retrieval of data.
  - File Systems: Splay trees can be used to store file system metadata, such as the allocation table, directory structure, and file attributes.
  - Data Compression: Splay trees can be used to compress data by identifying and encoding repeating patterns.
  - Text Processing: Splay trees can be used in text processing applications, such as spell-checkers, where words are stored in a splay tree for quick searching and retrieval.
  - Graph Algorithms: Splay trees can be used to implement graph algorithms, such as finding the shortest path in a weighted graph.

### Randomized BST (Treap)

Treap properties:
- each node contains two variables: key and priority:
  - __key values hold BST property__: left.key < parent.key < right.key
  - __priority values hold Min-Heap property__: parent.priority < child.priority

Construction:
-  To construct a Treap, each time insert a node, assign a priority generated as a real number that is uniformly random and independent between 0 and 1. The expected depth of any node will be $O(\log n)$.

Operations:
- __Search, insert, and delete operations are done in $O(height) = O(\log n)$ time__
- Seach: $O(\log n)$
  - follow the BST search algorithm
- Insert/delete: $O(\log n)$
  - for insert operation, follow BST insert steps. Then apply rotation operation until the node sits in the correct postion based on its property
  - delete is the inverse of insertion
- Split/merge: $O(\log n)$
  - spliting along pivot key value P: insert a new key with ke value P, and priority $-\infty$. This node with sits at root, then deleting the root will result in two subtrees.
  - merge: create a dummy node with key bigger than root.key of one sub-treap and smaller than the key.root of the other sub-treap. Then rotate the dummy node down to a leaf, and then cut it off.


## Skip-list

Skip-list is a sorted linked list with some random shortcuts.

Construct a skip-list:
- Recursivly add a short-cut layer to the sorted array
- For each node in the original list, repeatedly compute a binary value 0 or 1, with probability 0.5. If 1, then make a copy of the node in the new layer. The duplicates are stacked up in levels.
-  The depth of a skip list storing n keys is $O(\log n)$ with high probability

Operations:
- Seach: $O(\log n)$

## Algorithm UIUC 473

__Idea__: The basic idea is to pick a hash function h that maps every possible item $x$ to a small integer $h(x)$. Then we use the hash value $h(x)$ as a key to access $x$ in the data structure.

In its simplest form, a hash table is an array, in which each item x is stored at index $h(x)$.

__formal definition__: given a set of $n$ items from universe $\mathbb{U}$, and a hash table of size $m$, a hash function $h$ maps each possible item in $\mathbb{U}$ to a slot ($T[h(x)]$) in the hash table.
$$h: \mathbb{U} \rightarrow \{0, 1, \dots, m-1\}$$

__Direct access table (array)__ is a hash table with size = size of universe. Here the hash function could be defined as $h(x) = x$.

__Goal__: in many cases we are required to store a small subset of a elements in a universe, and as the result it is not efficient to have a table as big as the size of the universe. While using a smaller hash table saves space, the downside is the possibility of _collision_.

__Randomization__: it is not possible to define a perfect hash function, because we don't know the input values in advance (so we could define a hash function with no collision. The only way to avoid a worse-case (all inputs colide into the same spot in the hash table), we need to __define the hash function randomly__.
- Note: the goal is to have some level of randomness to guarantee good performance for the hash function.

---
__Define ideal hash function__:  Ideal random hash functions are k-uniform for every positive integer $k$
- __uniform__: . A family $\mathcal{H}$ of hash functions is uniform if choosing a hash function uniformly at random from $\mathcal{H}$ makes every hash value equally likely for every item in the universe. $$Pr[h(x) = i] = \frac{1}{m}$$
  - Note: this is not a useful property: we could define a family of hash function $const_a(x) = a ;\forall x$, which is useless!
- __universal__ (goal is to minimize collision): A family of hash functions is universal if, for any two items in the universe, the probability of collision is as small as possible.$$ \underset{h\in \mathcal{H}} {Pr} [h(x) = h(y)] \leq \frac{1}{m}\forall x\neq y$$
- __near universal__: A family of hash functions is near-universal if the probability of collision is close to idea.$$ \underset{h\in \mathcal{H}} {Pr} [h(x) = h(y)] \leq \frac{2}{m}\forall x\neq y$$
- __k-uniform (strongly k-universal)__: for any sequence of $k$ disjoint keys and any sequence of $k$ hash values, the probability that each key maps to the
corresponding hash value is $\frac{1}{m^k}$.

 $$\underset{h\in \mathcal{H}} {Pr}[h(x_1)==a_1, \dots, h(x_k)==a_k] = \frac{1}{m^k}$$
---
__Sollution to avoid collition__
1. __Chaining__: In a chained hash table, each entry $T[i]$ is a linked-list of all items that are hashed to $T[i]$.
  - If $h$ is choosen from a _universal_ set of hash function, the expected length of the largest linked-list of size $l(x)$ will be at most $\frac{n}{m}$, which is called the _load factor_ of a hash table ($\alpha$).
  - For a _near universal_ set of hash function the load factor will be $2\alpha$, and the expected time for an unsuccessful search in this case is $O(1+\alpha)$.
  - Note: if instead of a linked-list we employ BST, the expected time for a search will be $O(1+l(x))$. In the case of near universal hash function the expected time is $O(1+\log \alpha)$.
  - Note: if $\alpha = 1$, the expected size of the largest linked-list will be $\theta(\log n/ \log \log n)$
3. __Multiplicative hashing__: is a simple technique to achieve near-universal hashing.
  - notations:
    - __Salt__ parameter (a):  is chosen uniformly at random when the hash table is
created and remains fixed for the entire lifetime of the table.
    - for any non-negative integer $n$, $[n] = \[ 0,\dots, n-1 \]$ and  $[n]^+ = [ 1, \dots, n-1 ]$
    - $|\mathcal{U}| = 2^w$; elements in the universe are  w-bit words
    - $m = 2^l$; hash values are l-bit wrods
  - Methods:
    1. __Prime multiplicative hashing (_near-unversal_)__:
    
      For prime number $p> |\mathcal{U}|$ and any integer $a \in [p]^+$ (salt), prime multiplicative hashing is defined as: $\boxed{h_a(x) = (ax \mod p) \mod m}$

    2.  __Modified multiplicative hashing (universal)__:

      For any integers $a \in [p]^+$ and $b \in [p]$, the modified multiplicative hash function is defined as: $\boxed{h_{a,b}(x) = ((ax+b) \mod p) \mod m}$.
    3. __Binary multiplicative hashing (near universal)__:

       For any odd integer $a \in [2^w]$ (salt) the hash function is defined as $\boxed{hb_{a}(x) = \lfloor \frac{(ax) \mod 2^w}{2^{w-l}} \rfloor}$.

4. __Perfect Hashing__: use a hash table of size $m = n$ and a near-universal hash function, but instead of linked lists, employ a secondary hash tables to resolve collisions. In this method the $j^{th}$ secondary hash table has size $2n^2_j$, where $n_j$ is the number of items whose primary hash value is $j$.
  -  Here with probability at least 1/2, the secondary hash table has no collisions at all, so the worst-case search time in any secondary hash table is $O(1)$.
  - If we discover a collision in some secondary hash table, we can simply rebuild that table with a new near-universal hash function.
5. __Open addressing__:
  
  For a sequence of hash functions $(h_0, \dots, h_{m−1}>)$, such that for any item $x$, the probe sequence $〈h_0 (x),\dots, h_{m−1} (x))$ is a permutation of$(0, \dots, m − 1)$ (meaning different hash functions in the sequence always map $x$ to different locations in the hash table), the following algorithm returns
    - the array index $i$ if $T[i] = x$,
    - 'absent' if $x$ is not in the table but there is an empty slot, and
    - 'full' if $x$ is not in the table and there is no empty slots.

#### Hash

__Hash function: Overview__

Hash datastructure is used when we are dealing with dynamic data and require to have a fast lookup (search) operation. Hash data structure provides an _exprected_ constant running time for _insert, delete, and seach_ operations. However, the time complexity rely on how we define the hash function, which maps the input variable into index for stroing it in a hash table.

Some preliminary definision
- universe: a complete set of all variables, among which we need to store a subset of them with high probability.
- sample size: a subset of variables we want to store in our datastrucutre. Size of sample = $n$.
- hash table: a datastructure (array) that we store the input data into it by mapping the input into an index in the hash table. Size of hash table = $m$

__Collision__: although in concept hasing could be fast, in reality collision could reduce its performance to the point that it perform like an array. Collision occurs when hash function maps two or more input variables into the same bucket at the hash table. There are several methods addressing the collusion, among which are chaining and open-addressing.

__Chaining__

This method could solve the collision problem by denining a linked-list for each bucket of the hash table, if more than one input variable maps to it.

__Open-adressing__

Another soluton to collision s open-addressing, in which if collision occurs, we employ a function to find the next available free bucket in the hash table. Unlike chaining, implementation of open-addressing requires is more complex, unless we only want to have insert and lookup operations (no deletion).



#### Bloom filter

Furthermore, implementing a delete operation is more complicated than hash tables.

Some application of bloom filter are: spell check, forbidden password, and internet routers.

__What makes a good hash function__

A good hash function satisfies (approximately) the assumption of simple uniform hashing: each key is equally likely to hash to any of the $m$ slots, independently of where any other key has hashed to. Problem: we occationally know the distribution of hashed keys. The hash values needs to be independent of any patterns that might exist in the data.


- 2 properties of a good hash function:
  good distribution over hashtable and minimal colusion
- 2 challenges when developing a hash function: could cover a wide range of inputs, with minimal colusions

- Some of cryptographic hash functions
  - MD5 hash functions
  - SHA256

1. Division method: $$h(k) = k \mod m$.

  A prime not too close to an exact power of 2 is often a good choice for $m$ (size of hash table).

2. Multiplication method: $h(k) = \lfloor m (kA \mod 1)\rfloor, 0<A<1$

  Knuth suggest $A = (\sqrt {5} -1)\times 0.5$

  One advantage of this methid is the value of $m$ here is not critical.



__Ideas for preventing colusion__
- Avoid colision when the input is a list: _polynomially rolling hash function_

  The location of each element in the list here matters.
- Steps:
  - divide the input keys into 64/32 bit chuncks.
  - pick a prime number $p$ and it should not be a multiplier of $M$ (a large number)
  - $h(i) = (\sum chunck_i \times p^i) \mod M$

__Load factor:__ given $n$ inputs and hash table of size $m$, the load factor $\alpha = \frac{n}{m}$.

__Problem with a fixed hash function__: a knwn function is volnerable to collision, via reverse engineering colide outputs to come up with input data.


__Universal hash function__: randomly choose a hash function from a large family of hash functions.

  __Universal property__: if a hash function satisfies the following, it is called universal hash function: $$\forall k_i,k_j ; k_i \neq k_j: P_{H}(h(k_i)=h(k_j)) \leq \frac{1}{m}$$

  On average the size of the linked-list in case of colusion is $\alpha$, which makes the _average complexity_ of insert,delete, and find $≈ α$

 $$H_{a,b}(k) = (ak+b)\mod p)\mod m)$$
